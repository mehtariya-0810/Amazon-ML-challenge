```python
from sentence_transformers import SentenceTransformer

import numpy as np
import pandas as pd
import csv

import time
```

```python
train_path = "../input/amazon-ml-challenge/train.csv"
test_path = "../input/amazon-ml-challenge/test.csv"
train_df = pd.read_csv(train_path,escapechar="\\",quoting=csv.QUOTE_NONE)
#test_df = pd.read_csv(test_path,escapechar="\\",quoting=csv.QUOTE_NONE)
```

```python
model = SentenceTransformer('paraphrase-MiniLM-L3-v2',device='cuda')
start = time.time()
BATCH_SIZE = 4096
for i in range(0,len(train_df)-BATCH_SIZE,BATCH_SIZE):
    if i == 0:
        train_embed = model.encode(train_df["TITLE"][i:i+BATCH_SIZE].values,batch_size=BATCH_SIZE,show_progress_bar=False)
    else:
        tmp = model.encode(train_df["TITLE"][i:i+BATCH_SIZE].values,batch_size=BATCH_SIZE,show_progress_bar=False)
        train_embed = np.concatenate([train_embed,tmp])
    if(i % 50 == 0):
        print("Time taken : ",time.time() - start," sec")
        
np.save("./train_embed_2.npy",train_embed)
```
